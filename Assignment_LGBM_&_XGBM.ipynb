```python
# ============================================================
# 0.  SET‑UP & DATA LOAD
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# ML + tuning libs
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, classification_report,
                             roc_auc_score, RocCurveDisplay)

# Install once per Colab session
# ‑q = quiet; remove the flag if you want full install logs
# !pip install -q lightgbm xgboost

import lightgbm as lgb
from xgboost import XGBClassifier
import warnings, os, sys, json, math, random
warnings.filterwarnings("ignore", category=UserWarning)

# ------------ Load data ------------
train_df = pd.read_csv('Titanic_train.csv')
test_df  = pd.read_csv('Titanic_test.csv')
print("Train shape:", train_df.shape, "  Test shape:", test_df.shape)
display(train_df.head())

# ============================================================
# 1.  EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================

print("Missing values per column")
display(train_df.isna().sum())

# Histograms for numeric columns
num_cols = train_df.select_dtypes(include=['int64','float64']).columns
train_df[num_cols].hist(figsize=(12,8), bins=20, edgecolor="k")
plt.suptitle("Numeric Feature Distributions", y=1.03, fontsize=16)
plt.show()

# Box‑plots
plt.figure(figsize=(10,5))
sns.boxplot(data=train_df[num_cols.drop('Survived')])
plt.title("Box‑Plot: Detecting Outliers")
plt.xticks(rotation=45)
plt.show()

# Bar plots: Survival vs categorical
cat_cols = ['Sex','Pclass','Embarked']
for col in cat_cols:
    plt.figure(figsize=(4,3))
    sns.barplot(x=col, y='Survived', data=train_df, ci=None)
    plt.title(f"Survival Rate by {col}")
    plt.ylim(0,1)
    plt.show()

# Scatter: Age vs Fare coloured by survival
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train_df, alpha=0.7)
plt.title("Age vs Fare coloured by Survival")
plt.show()

# ============================================================
# 2.  DATA PRE‑PROCESSING
# ============================================================
target = 'Survived'
X = train_df.drop(columns=[target])
y = train_df[target]

# --- Column lists
numeric_features = X.select_dtypes(include=['int64','float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# --- Pre‑processing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocess = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
])

# --- Train / Test split
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y)

# Helper: evaluation function
def evaluate(model, X_tr, y_tr, X_val, y_val, name="model"):
    y_pred = model.predict(X_val)
    acc  = accuracy_score(y_val, y_pred)
    prec = precision_score(y_val, y_pred)
    rec  = recall_score(y_val, y_pred)
    f1   = f1_score(y_val, y_pred)
    auc  = roc_auc_score(y_val, model.predict_proba(X_val)[:,1])
    print(f"\n{name} Metrics")
    print("─"*30)
    print(f"Accuracy : {acc:0.4f}")
    print(f"Precision: {prec:0.4f}")
    print(f"Recall   : {rec:0.4f}")
    print(f"F1‑score : {f1:0.4f}")
    print(f"ROC‑AUC  : {auc:0.4f}")
    RocCurveDisplay.from_estimator(model, X_val, y_val)
    plt.title(f"ROC Curve – {name}")
    plt.show()
    return {'Model':name, 'Accuracy':acc, 'Precision':prec,
            'Recall':rec, 'F1':f1, 'ROC‑AUC':auc}

# ============================================================
# 3‑A.  LightGBM  (with quick GridSearch)
# ============================================================
lgb_pipe = Pipeline(steps=[
    ('preprocess', preprocess),
    ('classifier', lgb.LGBMClassifier(
        objective='binary', random_state=42, n_jobs=-1))
])

lgb_param_grid = {
    'classifier__n_estimators':[200,400],
    'classifier__max_depth'   :[-1, 6, 10],
    'classifier__learning_rate':[0.03,0.1]
}

lgb_grid = GridSearchCV(
    lgb_pipe, lgb_param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring='f1', n_jobs=-1, verbose=0)

lgb_grid.fit(X_train, y_train)
print("Best LGBM params:", lgb_grid.best_params_)

lgb_best = lgb_grid.best_estimator_

# ============================================================
# 3‑B.  XGBoost  (with quick GridSearch)
# ============================================================
xgb_pipe = Pipeline(steps=[
    ('preprocess', preprocess),
    ('classifier', XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        random_state=42,
        n_jobs=-1,
        use_label_encoder=False))
])

xgb_param_grid = {
    'classifier__n_estimators':[300,500],
    'classifier__max_depth':[3,5],
    'classifier__learning_rate':[0.03,0.1],
    'classifier__subsample':[0.8,1.0]
}

xgb_grid = GridSearchCV(
    xgb_pipe, xgb_param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring='f1', n_jobs=-1, verbose=0)

xgb_grid.fit(X_train, y_train)
print("Best XGB params:", xgb_grid.best_params_)

xgb_best = xgb_grid.best_estimator_

# ============================================================
# 4.  EVALUATION & COMPARISON
# ============================================================
results = []
results.append(evaluate(lgb_best, X_train, y_train, X_valid, y_valid,
                        name="LightGBM"))
results.append(evaluate(xgb_best, X_train, y_train, X_valid, y_valid,
                        name="XGBoost"))

pd.DataFrame(results).set_index('Model').style.background_gradient(cmap='Blues')

# ============================================================
# 5.  TRAIN ON FULL TRAINING DATA & PREDICT TEST SET  (optional)
# ============================================================
# If you want to generate predictions for Kaggle‑style submission:

lgb_best.fit(X, y)
test_pred_lgb = lgb_best.predict(test_df)
pd.DataFrame({'PassengerId': test_df['PassengerId'],
              'Survived': test_pred_lgb}).to_csv('lgb_submission.csv', index=False)

xgb_best.fit(X, y)
test_pred_xgb = xgb_best.predict(test_df)
pd.DataFrame({'PassengerId': test_df['PassengerId'],
              'Survived': test_pred_xgb}).to_csv('xgb_submission.csv', index=False)

print("Submission files saved: lgb_submission.csv, xgb_submission.csv")
```
